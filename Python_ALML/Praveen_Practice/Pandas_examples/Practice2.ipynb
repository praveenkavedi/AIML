```python
import pandas as pd

movie = pd.read_csv('movie.csv')
movie
```

```python
rating = pd.read_csv('ratings.csv')
rating
```

```python
df_merge = movie.merge(rating,on = 'movie id', how='inner')
df_merge
```

```python
avg_rating = df_merge.groupby(['movie title'])[['rating']].mean()
avg_rating
```

```python
avg_rating.rename(columns={'rating':'Avg_rating'}, inplace=True)
avg_rating
```

```python
avg_rating.sort_values(by='Avg_rating')
```

```python
avg_rating[avg_rating['Avg_rating']==5]
```

```python
avg_rating['Avg_rating']
```

```python
df_merge
```

```python
movie_count = df_merge.groupby(['movie title'])[['rating']].count().reset_index()
movie_count.sort_values(by=['rating'],ascending=False)
```

```python
movie_count.rename(columns={'rating':'Rating_counts'}, inplace=True)
movie_count.sort_values(by=['Rating_counts'], ascending=False)
```

```python
movie_100=movie_count[movie_count['Rating_counts']>100]
movie_100
```

```python
top_movies =  movie_100.sort_values(by=['Rating_counts'], ascending=False)
top_movies
```

```python
avg_rating.head()
```

```python
top_movies.head()
```

```python
top_rated = avg_rating.merge(top_movies, on='movie title', how='inner')
top_rated
```
